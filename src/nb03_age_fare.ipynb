Imputed Age vs. Fare Per Person
===============================

This notebook will explore the relationship between two engineered features in the Titanic dataset: age (imputed) and total fare divided by the group size. The goal is to find a simplified between these features, as well as the gender of each passenger.

```python
# set up code and dataframe with training data
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.colors as cls

datestamp = "20250129.150949"
df_ = pd.read_csv(f"../data/kaggle/train.clean.{datestamp}.csv")
print(df_.shape)
```

```python
# set up the X, y sets for analysis
x1_colnames = [ 'AgeImputed', 'FarePerPerson' ]
x2_colnames = [ 'SexOrd' ]
y_colnames = [ 'Survived' ]

# get SexOrd subsets for markers
df_died     = df_[ df_[ 'Survived' ] == 0 ]
df_survived = df_[ df_[ 'Survived' ] == 1 ]

# get the X and y values for the male passengers
X_died = df_died[ x1_colnames ]
y_died = df_died[ x2_colnames ].values

# get the X and y values for the male passengers
X_survived = df_survived[ x1_colnames ]
y_survived = df_survived[ x2_colnames ].values

df_[["Name", "AgeImputed", "SexOrd", "FarePerPerson", "Survived"]].head(5)
```

```python
# Create the colormap
cmap = cls.ListedColormap(['blue', 'orange' ])

# scatter plot
plt.figure(figsize=(12,6))
plt.scatter( X_survived[x1_colnames[0]], X_survived[x1_colnames[1]], c=y_survived, cmap=cmap, s=8, marker="o" )
plt.scatter( X_died[x1_colnames[0]], X_died[x1_colnames[1]], c=y_died, cmap=cmap, s=10, marker="x" )

plt.xlabel( 'AgeImputed' )
plt.ylabel( 'FarePerPerson' )

#plt.xlim()
plt.ylim(-2,62)

plt.title(f"{x1_colnames[0]} vs. {x1_colnames[1]}")
plt.show()
```
